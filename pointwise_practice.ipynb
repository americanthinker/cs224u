```python
import re
from collections import Counter
from typing import List, Tuple
import numpy as np
```

```python
pattern = '\w+'
text = ' '.join(re.findall(pattern ,text))
```

```python
vocab = set(text.lower().split())
```

```python
words = text.lower().split()
```

```python
total_words = len(words)
```

```python
counts = Counter(words)
```

```python
probs = {k : v/total_words for k,v in counts.items()}
```

```python
probs['las']
```

```python
probs['vegas']
```

```python
def ngram_generator(string: str, n: int) -> List[Tuple[str]]:
    '''
    Given a sentence, returns a list of tuple n-grams. 
    '''
    words = []
    pattern = '\w+'
    answer = re.findall(pattern, string) 
    length = len(answer)
    if length < n:
        n = length
    for index, _ in enumerate(answer):
        words.append(tuple(answer[index:index + n]))
        if index == length - n:
            break
    return words
```

```python
bigrams = ngram_generator(text.lower(), 2)
```

```python
bigram_counts = Counter(bigrams)
```

```python
bigram_probs = {k: v/536 for k, v in bigram_counts.items()}
```

```python
bigram_probs[('las', 'vegas')]
```

```python
def pmi(a,b):
    numerator = bigram_probs[(a,b)]
    denomintaor = probs[a] * probs[b]
    return np.log(numerator/denomintaor)
```

```python
scores = []
for bigram in bigrams:
    score = pmi(bigram[0], bigram[1])
    scores.append((bigram, score))
```
